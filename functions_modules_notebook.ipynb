## User Defined Functions

```python
def printme(name):
    print("Hello", name)

printme("MIT")
```

```python
def printinfo(name, location='kothrud'):
    print(name, location)

printinfo("MIT")
printinfo("SNDT", "karveroad")
```

```python
def printinfo(arg1, *n):
    print(arg1)
    for i in n:
        print(i)

printinfo(10)
printinfo(70,60,50,40)
```

## Count Upper & Lower Case

```python
def check_string(s):
    upper = sum(1 for c in s if c.isupper())
    lower = sum(1 for c in s if c.islower())
    print(upper, lower)

check_string("MITpoly")
```

## Creating Modules

```python
# module1.py
def check_prime(num):
    if num <= 1:
        print("Not prime")
        return
    for i in range(2, num):
        if num % i == 0:
            print("Not prime")
            return
    print("Prime")

def factorial(n):
    f = 1
    for i in range(1,n+1):
        f *= i
    print(f)
```

```python
# module2.py
def add(a,b): return a+b
def sub(a,b): return a-b
def mul(a,b): return a*b
def div(a,b): return a/b
```

## Using Modules

```python
import module1 as m
m.check_prime(23)
m.factorial(5)
```

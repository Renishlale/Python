## ATM Class Example

```python
class ATM:
    def __init__(self):
        self.pin = None
        self.balance = 0
        print("ATM system initialized!")
        self.menu()

    def menu(self):
        print("Hello! How can I help you?")
        print("1. Set PIN")
        print("2. Change PIN")
        print("3. Check Balance")
        print("4. Withdraw Cash")

atm1 = ATM()
```

## Addition Class

```python
class Addition:
    def __init__(self):
        self.num1 = 1000
        self.num2 = 2000
        self.num3 = 3000

    def result(self):
        X = 10
        self.num = self.num1 + self.num2 + self.num3 + X
        print("Output", self.num)

Sum = Addition()
Sum.result()
```

## Class Variable Example

```python
class Demo:
    x = 100

a = Demo()
b = Demo()

print(a.x)
print(b.x)
```

## Parameterized Constructor

```python
class Student:
    def __init__(self, name, ids, college):
        print("Parameterized constructor")
        self.name = name
        self.ids = ids
        self.college = college

    def display(self):
        print(self.name, self.ids, self.college)

student = Student("JOHN", 2023, "MIT")
student.display()
```

## Bank Account Example

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def display(self):
        print(self.name, self.balance)

acc = BankAccount("Sneha", 10000)
acc.deposit(2000)
acc.withdraw(1500)
acc.display()
```

## Single Inheritance

```python
class Parent:
    def func1(self):
        print("Parent class")

class Child(Parent):
    def func2(self):
        print("Child class")

obj = Child()
obj.func1()
obj.func2()
```

## Multiple Inheritance

```python
class Mother:
    mothername = ""

class Father:
    fathername = ""

class Son(Mother, Father):
    def show(self):
        print(self.mothername, self.fathername)

s = Son()
s.mothername = "SITA"
s.fathername = "RAM"
s.show()
```

## Method Overriding

```python
class Class1:
    def m(self):
        print("Class1")

class Class2(Class1):
    def m(self):
        print("Class2")

obj = Class2()
obj.m()
```

## Multilevel Inheritance

```python
class Grandfather:
    def __init__(self, g):
        self.g = g

class Father(Grandfather):
    def __init__(self, f, g):
        self.f = f
        super().__init__(g)

class Son(Father):
    def __init__(self, s, f, g):
        self.s = s
        super().__init__(f, g)

    def show(self):
        print(self.s, self.f, self.g)

obj = Son("ABC", "MUKESH", "DHIRUBHAI")
obj.show()
```

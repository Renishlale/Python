```python
import os
os.makedirs('my_package', exist_ok=True)
open('my_package/__init__.py','w').close()
```

```python
with open('my_package/module1.py','w') as f:
 f.write('''def check_prime(num):
 if num==1:
  print("Not prime")
 elif num in (2,3):
  print("Prime number")
 elif num%2==0 or num%3==0:
  print("Not prime")
 else:
  print("Prime number")


def calculate_factorial(num):
 fact=1
 for i in range(1,num+1):
  fact*=i
 print("Factorial:",fact)
''')
```

```python
with open('my_package/module2.py','w') as f:
 f.write('''def add(a,b): return a+b

def mul(a,b): return a*b
''')
```

```python
from my_package import module1 as m1
m1.check_prime(13)
m1.calculate_factorial(13)
```

```python
from my_package import module2 as m2
print('Addition:', m2.add(2,4))
print('Multiplication:', m2.mul(2,4))
```

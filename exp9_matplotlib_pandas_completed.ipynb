```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
 'Name':['Aanya','Riya','Kabir','Arjun'],
 'Maths':[85,90,78,88],
 'Science':[92,87,80,85],
 'English':[88,91,75,82]
}
df=pd.DataFrame(data)
print('Full DataFrame:')
print(df)
```

```python
print(df[['Name','Maths']])
```

```python
df.plot(x='Name',y='Maths',kind='bar')
plt.title('Maths Marks of Students')
plt.show()
```

```python
data={'Name':['Amit','Sneha','Ravi','Priya'],'Age':[20,21,19,22],'Marks':[85,90,78,88]}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.head())
```

```python
print(df['Name'])
```

```python
print(df.shape)
print(df.columns)
```

```python
plt.plot(df['Name'],df['Marks'])
plt.title('Marks Trend')
plt.show()
```

```python
plt.bar(df['Name'],df['Marks'])
plt.title('Marks Comparison')
plt.show()
```

```python
plt.pie(df['Marks'],labels=df['Name'],autopct='%1.1f%%')
plt.title('Marks Distribution')
plt.show()
```

```python
plt.scatter(df['Age'],df['Marks'])
plt.xlabel('Age')
plt.ylabel('Marks')
plt.title('Age vs Marks')
plt.show()
```
